```python
# skeleton code to produce a 3D scatter plot of XYZ
import numpy as np
import pylab as plt
from mpl_toolkits . mplot3d import Axes3D
import astropy.units as u
from astropy.coordinates import SkyCoord
from astropy.coordinates import Galactic
from IPython.display import display

%cd C:\Users\titan\OneDrive\Desktop\McMaster\Intro Astro\A1
```

```python
#import data from csv file
data = np.loadtxt("TWA_astrometry.csv", delimiter=",", skiprows=1)  # skip header row if needed
```

```python
data_ra = data[:,0]*u.deg
data_dec = data[:,1]*u.deg
data_pm_ra = np.array(data[:,2])*u.mas/u.yr
data_pm_dec = np.array(data[:,3])*u.mas/u.yr
data_parallax = data[:,4]*u.mas
data_rad_v = data[:,5]*u.km /u.s
```

```python
data_distance = 1*u.arcsec / data_parallax.to(u.arcsec) *u.pc
```

```python
c_icrs = SkyCoord(
    ra = data_ra, 
    dec = data_dec,  
    distance = data_distance,
    pm_ra_cosdec = data_pm_ra,
    pm_dec = data_pm_dec,
    radial_velocity = data_rad_v,
    frame="icrs")
```

```python
# Position in Galactic XYZ
c_gal = c_icrs.transform_to(Galactic)
X = c_gal.cartesian.x
Y = c_gal.cartesian.y
Z = c_gal.cartesian.z

# Velocity in Galactic UVW
uvw = c_icrs.transform_to(Galactic).velocity.d_xyz
U, V, W = uvw.to('km/s')
```

```python
fig1 = plt.figure(figsize=(12, 6))

# plot XYZ positions
ax1 = fig1.add_subplot(1, 2, 1, projection='3d')
ax1.scatter(X , Y , Z, alpha=0.4)

# choose a random star
index = np . random . choice ( range (len ( X ) ) )

x = np.array([X[index].value])
y = np.array([Y[index].value])
z = np.array([Z[index].value])

# hardcode the axes limits ( will be helpful later )
ax1. set_xlim3d ( ax1. get_xlim3d () )
ax1. set_ylim3d ( ax1. get_ylim3d () )
ax1. set_zlim3d ( ax1. get_zlim3d () )

# plot 2d plane projections as plus signs
ax1. scatter (X , Y , np . ones_like ( Z ) * ax1. get_zlim3d () [0] , c ='b', marker ='+')
ax1. scatter (X , np . ones_like ( Y ) * ax1. get_ylim3d () [1] , Z , c ='g', marker ='+')
ax1. scatter ( np . ones_like ( X ) * ax1. get_xlim3d () [0] , Y , Z , c ='r', marker ='+')

# plot ‘stems ' to highlight the star 's position in each 2D projection
ax1. stem (x,y,z, orientation ='x', linefmt='r:', bottom = ax1. get_xlim3d () [0])
ax1. stem (x,y,z , orientation ='y', linefmt='g:', bottom = ax1. get_ylim3d () [1])
ax1. stem (x,y,z, orientation ='z', linefmt='b:', bottom = ax1. get_zlim3d () [0])

#hightlight the chosen star
ax1.scatter ( x, y, z, marker = '*', s=350, facecolor='white', edgecolor='black')

ax1. set_xlabel ('X [pc]') , ax1. set_ylabel ('Y [pc]') , ax1. set_zlabel ('Z [pc]')

#SECOND PLOT

ax2 = fig1.add_subplot(1, 2, 2, projection='3d')
ax2.scatter(U , V , W, alpha=0.4)

# choose the same star as before

u_star = np.array([U[index].value])
v_star = np.array([V[index].value])
w_star = np.array([W[index].value])

# hardcode the axes limits ( will be helpful later )
ax2. set_xlim3d ( ax2. get_xlim3d () )
ax2. set_ylim3d ( ax2. get_ylim3d () )
ax2. set_zlim3d ( ax2. get_zlim3d () )

# plot 2d plane projections as plus signs
ax2. scatter (U , V , np . ones_like ( W ) * ax2. get_zlim3d () [0] , c ='b', marker ='+')
ax2. scatter (U , np . ones_like ( V ) * ax2. get_ylim3d () [1] , W , c ='g', marker ='+')
ax2. scatter ( np . ones_like ( U ) * ax2. get_xlim3d () [0] , V , W , c ='r', marker ='+')

# plot ‘stems ' to highlight the star 's position in each 2D projection
ax2. stem (u_star,v_star,w_star, orientation ='x', linefmt='r:', bottom = ax2. get_xlim3d () [0])
ax2. stem (u_star,v_star,w_star, orientation ='y', linefmt='g:', bottom = ax2. get_ylim3d () [1])
ax2. stem (u_star,v_star,w_star, orientation ='z', linefmt='b:', bottom = ax2. get_zlim3d () [0])

#hightlight the chosen star
ax2.scatter (u_star,v_star,w_star, marker = '*', s=350, facecolor='white', edgecolor='black')

ax2. set_xlabel ('U [km/s]') , ax2. set_ylabel ('V [km/s]') , ax2. set_zlabel ('Z [km/s]')
fig1.savefig('figure1.png')
```
